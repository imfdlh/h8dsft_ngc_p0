# Create SQLite Function

## Connection Function

```python
def create_connection(path):
    connection = None
    try:
        connection = sqlite.connect(path)
        print("Connection to SQLite DB successful.")
    except Error as e:
        print(f"The error '{e}' occured.")
    return connection
```

## Command Function

```python
def execute_command(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
#         print("Command query executed successfully.")
    except Error as e:
        print(f"The error '{e}' occured.")
```

## Assessments

```python
assessments.info()
```

```python
create_assessments_tab = """
CREATE TABLE IF NOT EXISTS assessments (
    code_module TEXT,
    code_presentation TEXT,
    id_assessment INT,
    assessment_type TEXT,
    date FLOAT,
    weight FLOAT
)
"""
execute_command(oulad_students_conn, create_assessments_tab)
```

```python
assessments.isnull().any()
```

```python
for i in range(len(assessments)):
    insert_table = f"""
    INSERT INTO assessments (
        {", ".join(assessments.columns.to_list())}
    )
    VALUES (
        '{assessments['code_module'][i]}', '{assessments['code_presentation'][i]}',
        {assessments['id_assessment'][i]}, '{assessments['assessment_type'][i]}',
        {'NULL' if np.isnan(assessments['date'][i]) else assessments['date'][i]}, {assessments['weight'][i]}
    )
    """
    execute_command(oulad_students_conn, insert_table)
```

```python
pd.read_sql_query("SELECT * FROM assessments", oulad_students_conn)
```

# Select Query

```python
result = pd.read_sql_query(
    """
    SELECT
        si.id_student, si.region, si.code_module,
        a.assessment_type, sa.score
    FROM student_info as si
    INNER JOIN student_assessment as sa
        ON si.id_student = sa.id_student
    INNER JOIN assessments as a
        ON sa.id_assessment = a.id_assessment
    WHERE si.code_presentation = '2014B'
    """,
    oulad_students_conn
)
result
```
